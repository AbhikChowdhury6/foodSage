```python
# Update DataFrame with safety and type columns
df['USA_safe'] = df['Status'].map({'red': 0, 'yellow': 1, 'green': 1})
df['USA'] = df['Status'].map({'red': 0, 'yellow': 1, 'green': 1})
df['EU_safe'] = df['Status'].map({'red': 0, 'yellow': 0, 'green': 1})
df['EU'] = df['Status'].map({'red': 0, 'yellow': 0, 'green': 1})

# Add ingredient type categorization
def categorize_ingredient(name):
    name = name.lower()
    if any(x in name for x in ['acid', 'citrate', 'acetate', 'lactate']):
        return 'Acid/Salt'
    elif any(x in name for x in ['vitamin', 'mineral', 'iron', 'calcium']):
        return 'Vitamin/Mineral'
    elif any(x in name for x in ['gum', 'starch', 'fiber']):
        return 'Thickener'
    elif any(x in name for x in ['extract', 'powder', 'natural']):
        return 'Natural Extract'
    elif any(x in name for x in ['color', 'dye', 'red', 'yellow', 'blue']):
        return 'Colorant'
    else:
        return 'Other'

df['Type'] = df['Ingredient'].apply(categorize_ingredient)

# Generate summary statistics
summary = pd.DataFrame({
    'Total_Count': len(df),
    'USA_Approved': df['USA'].sum(),
    'EU_Approved': df['EU'].sum(),
    'Banned_Both': len(df[df['Status'] == 'red']),
    'Type_Distribution': df['Type'].value_counts().to_dict()
})

print("\nIngredient Safety Analysis:")
print(f"Total Ingredients: {len(df)}")
print(f"USA Approved: {df['USA'].sum()} ({df['USA'].mean()*100:.1f}%)")
print(f"EU Approved: {df['EU'].sum()} ({df['EU'].mean()*100:.1f}%)")
print("\nType Distribution:")
print(df['Type'].value_counts())

# Display enhanced DataFrame
print("\nEnhanced DataFrame Sample:")
print(df[['Ingredient', 'Status', 'Type', 'USA_safe', 'EU_safe']].head())
```

```python
import pandas as pd

def parse_test_file(filepath):
    with open(filepath, 'r') as file:
        content = file.read()
    
    # Split into product sections
    products = []
    sections = content.split('\n#')
    
    for section in sections:
        if section.strip():
            # Parse product name and ingredients list
            lines = section.strip().split('\n')
            product_name = lines[0].replace('#', '').strip()
            
            # Extract ingredients, cleaning up the format
            ingredients = []
            for line in lines[1:]:
                if 'test_' in line:  # Skip test_N = [ lines
                    continue
                # Clean up ingredients
                ingredients.extend([i.strip(' "[],"') for i in line.split('",') if i.strip(' "[],')])
            
            # Create image filename (simplified version)
            image_name = product_name.lower().replace(' ', '_').replace('-', '_') + '.jpg'
            
            products.append({
                'product_name': product_name,
                'image_name': image_name,
                'ingredients': ingredients
            })
    
    return pd.DataFrame(products)

# Create DataFrame
products = parse_test_file('/Users/wgm/dockerStation/docker-images/Vizai/VizAi/Image_Names_Ingredients/TestSet_Names_Ingredients.txt')

# Transform the DataFrame to have columns: product, imagefile, ingredient 1 - n
max_ingredients = products['ingredients'].apply(len).max()

# Create new DataFrame with expanded columns
expanded_columns = ['product_name', 'image_name'] + [f'ingredient_{i+1}' for i in range(max_ingredients)]
expanded_df = pd.DataFrame(columns=expanded_columns)

for idx, row in products.iterrows():
    row_data = [row['product_name'], row['image_name']] + row['ingredients'] + [None] * (max_ingredients - len(row['ingredients']))
    expanded_df.loc[idx] = row_data

# Display results
print("Dataset Overview:")
print(f"Total products: {len(expanded_df)}")
print("\nSample Product Entry:")
print(expanded_df.head())
```

```python
from PIL import Image as PILImage
import openpyxl
from openpyxl.drawing.image import Image
import os

def create_thumbnail(image_path, size=(10, 10)):
    """Create a thumbnail of specified size"""
    try:
        with PILImage.open(image_path) as img:
            # Convert to RGB if needed
            if img.mode != 'RGB':
                img = img.convert('RGB')
            # Create thumbnail
            img.thumbnail(size)
            # Save thumbnail with _thumb suffix
            thumb_path = image_path.rsplit('.', 1)[0] + '_thumb.png'
            img.save(thumb_path, 'PNG')
            return thumb_path
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None

expanded_df = expanded_df.fillna('None')
print(expanded_df.head())

# Save the expanded_df to Excel
excel_filename = 'expanded_df_with_images.xlsx'
expanded_df.to_excel(excel_filename, index=False)

# Load workbook
wb = openpyxl.load_workbook(excel_filename)
ws = wb.active

# Process images and add thumbnails
for idx, row in expanded_df.iterrows():
    # Original image path
    image_path = f"/Users/wgm/dockerStation/docker-images/Vizai/VizAi/Image_Names_Ingredients/test_{idx + 1}.png"
    
    if os.path.exists(image_path):
        # Create thumbnail
        thumb_path = create_thumbnail(image_path)
        if thumb_path:
            # Add thumbnail to Excel
            img = Image(thumb_path)
            # Adjust cell width/height for thumbnails
            ws.row_dimensions[idx + 2].height = 15
            ws.column_dimensions['B'].width = 15
            img.anchor = f'B{idx + 2}'
            ws.add_image(img)



# Save workbook
wb.save(excel_filename)

# Clean up thumbnails (optional)
for idx in range(len(expanded_df)):
    thumb_path = f"/Users/wgm/dockerStation/docker-images/Vizai/VizAi/Image_Names_Ingredients/test_{idx + 1}_thumb.png"
    if os.path.exists(thumb_path):
        os.remove(thumb_path)
```

```python
def analyze_ingredient_safety(expanded_df, safety_df):
    # Create lookup dictionary for ingredient status
    safety_lookup = dict(zip(safety_df['Ingredient'].str.lower(), safety_df['Status']))
    
    # Initialize counters
    total_red = 0
    total_yellow = 0
    total_green = 0
    unknown = 0
    
    # Get all ingredient columns
    ingredient_cols = [col for col in expanded_df.columns if col.startswith('ingredient_')]
    
    # Check each ingredient in each product
    for _, row in expanded_df.iterrows():
        for col in ingredient_cols:
            ingredient = row[col]
            if pd.isna(ingredient) or ingredient is None:
                continue
                
            # Look up ingredient status (case-insensitive)
            status = safety_lookup.get(ingredient.lower())
            
            if status == 'red':
                total_red += 1
            elif status == 'yellow':
                total_yellow += 1
            elif status == 'green':
                total_green += 1
            else:
                unknown += 1
    
    print(f"Safety Analysis of Product Ingredients:")
    print(f"Banned (Red) ingredients: {total_red}")
    print(f"Caution (Yellow) ingredients: {total_yellow}")
    print(f"Safe (Green) ingredients: {total_green}")
    print(f"Unknown ingredients: {unknown}")

# Run analysis
analyze_ingredient_safety(expanded_df, df)
```

```python
import pandas as pd

def check_ingredient_safety(ingredient, safety_df):
    """Check safety status of ingredient"""
    # Case insensitive search
    match = safety_df[safety_df['Ingredient'].str.lower() == ingredient.lower()]
    if not match.empty:
        return match.iloc[0]['Status']
    return 'unknown'

def analyze_product_safety(product_df, safety_df):
    """Analyze safety of each product"""
    product_analysis = []
    
    # Analyze each product
    for _, row in product_df.iterrows():
        ingredients = [col for col in row.index if col.startswith('ingredient_')]
        ingredient_counts = {'red': 0, 'yellow': 0, 'green': 0, 'unknown': 0}
        
        # Check each ingredient
        for ing_col in ingredients:
            ingredient = row[ing_col]
            if pd.notna(ingredient):
                status = check_ingredient_safety(ingredient, safety_df)
                ingredient_counts[status] += 1
        
        # Classify product safety
        if ingredient_counts['red'] > 0:
            overall_status = 'red'
        elif ingredient_counts['yellow'] > 0:
            overall_status = 'yellow'
        elif ingredient_counts['green'] > 0:
            overall_status = 'green'
        else:
            overall_status = 'unknown'
            
        product_analysis.append({
            'product_name': row['product_name'],
            'red_ingredients': ingredient_counts['red'],
            'yellow_ingredients': ingredient_counts['yellow'],
            'green_ingredients': ingredient_counts['green'],
            'unknown_ingredients': ingredient_counts['unknown'],
            'overall_status': overall_status
        })
    
    return pd.DataFrame(product_analysis)

# Analyze products
results = analyze_product_safety(expanded_df, df)

# Print summary
print("\nProduct Safety Analysis:")
print("\nOverall Product Classification:")
print(results['overall_status'].value_counts())

print("\nIngredient Counts:")
print(f"Total Red Ingredients: {results['red_ingredients'].sum()}")
print(f"Total Yellow Ingredients: {results['yellow_ingredients'].sum()}")
print(f"Total Green Ingredients: {results['green_ingredients'].sum()}")
print(f"Total Unknown Ingredients: {results['unknown_ingredients'].sum()}")

print("\nProducts by Safety Level:")
print("\nRed Products (Contains banned ingredients):")
print(results[results['overall_status'] == 'red']['product_name'].tolist())
print("\nYellow Products (Contains cautionary ingredients):")
print(results[results['overall_status'] == 'yellow']['product_name'].tolist())
print("\nGreen Products (Contains only safe ingredients):")
print(results[results['overall_status'] == 'green']['product_name'].tolist())

```
